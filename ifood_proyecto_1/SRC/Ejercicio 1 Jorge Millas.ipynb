```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('../data/ml_project1_data.csv')
# Calcular el gasto total en iFood
data['Total_Spent'] = (data['MntWines'] + data['MntFruits'] + data['MntMeatProducts'] + data['MntFishProducts'] + data['MntSweetProducts'] + data['MntGoldProds'])
# Seleccionar las columnas relevantes
income_spent = data[['Income', 'Total_Spent']]
plt.figure(figsize=(10, 6))
plt.scatter(income_spent['Income'], income_spent['Total_Spent'], alpha=0.5)
plt.title('Relación entre Ingresos Anuales y Gasto Total en iFood')
plt.xlabel('Ingresos Anuales')
plt.ylabel('Gasto Total en iFood')
plt.grid(True)
plt.show()
```

Visualizando la gráfica, quitando los outliers que hacen que se represente datos fuera de lo normal, se observa una correlación positiva entre los ingresos anuales y la cantidad de dinero que gastan en iFood, es decir, cuanto mas ingresos mas dinero gastado en iFood.
