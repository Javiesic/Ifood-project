```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos
data = pd.read_csv('../data/ml_project1_data.csv')

# Filtrar los datos por estado civil
solteros = data[data['Marital_Status'] == 'Single']
casados = data[data['Marital_Status'] == 'Married']

solteros['Total_Spent'] = (solteros['MntWines'] + solteros['MntFruits'] + solteros['MntMeatProducts'] + solteros['MntFishProducts'] + solteros['MntSweetProducts'] + solteros['MntGoldProds'])

casados['Total_Spent'] = (casados['MntWines'] + casados['MntFruits'] + casados['MntMeatProducts'] + casados['MntFishProducts'] + casados['MntSweetProducts'] + casados['MntGoldProds'])

# Calcular el gasto promedio
gasto_promedio_solteros = round(solteros['Total_Spent'].mean(), 2)
gasto_promedio_casados = round(casados['Total_Spent'].mean(), 2)

print(f"Gasto promedio de solteros: {gasto_promedio_solteros}")
print(f"Gasto promedio de casados: {gasto_promedio_casados}")

plt.figure(figsize=(10, 6))
sns.barplot(x='Marital_Status', y='Total_Spent', data=pd.concat([solteros, casados]), hue='Marital_Status')
plt.title('Gastos Totales por Estado Civil')
plt.xlabel('Estado Civil')
plt.ylabel('Gasto Total')
plt.show()




```

La gráfica muestra que el grupo de los solteros tiene un promedio de gasto más elevado que los casados, por lo que segun la gráfica se puede concluir que el grupo de los solteros gasta más dinero que el grupo de los casados
