```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
ruta = "C:/Users/adria/OneDrive/Escritorio/ESIC 24-25/P. Proyecto Prediccion/ifood_proyecto_1/data/ml_project1_data/ml_project1_data.csv"
data = pd.read_csv(ruta)
data.info()
```

```python
# Definir categoras de ingresos
rangos_ingresos = [0, 15000 ,30000, 45000, 70000, float('inf')]
rangos = ['Bajos', 'Medio-Bajo', 'Medios', 'Medio-Alto','Altos']
data['Categoria_Ingresos'] = pd.cut(data['Income'], bins=rangos_ingresos, labels=rangos)

# Calcular el gasto promedio en cada categora de productos (a travs de groupby)
gasto_por_categoria = data.groupby('Categoria_Ingresos')[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].mean()

# Visualizar los datos (grfico de barras)
print(gasto_por_categoria)
gasto_por_categoria.plot(kind='bar', figsize=(12, 8))
plt.title('Gasto Promedio en Productos por Categora de Ingresos')
plt.xlabel('Categora de Ingresos')
plt.ylabel('Gasto Promedio')
plt.legend(title='Productos')
plt.show()
```
